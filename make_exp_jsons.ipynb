```python
import numpy as np
import json
import matplotlib.pyplot as plt
import numpy.random as nr
```

```python
# create the feature_probabilities (want to check these for auto_correlation...)
n_trials = 10;
n_states = 3
n_features = 3

state_feature_probs = np.zeros((n_states,n_features,n_trials))
state_feature_probs[:,:,0] = .25 + .5*nr.rand(n_states,n_features)

for i in range(1,n_trials):
    this_trial_sf = state_feature_probs[:,:,i-1] + nr.normal(0,.025,size = (3,3))
    this_trial_sf[this_trial_sf < .25] = .25
    this_trial_sf[this_trial_sf > .75] = .75
    state_feature_probs[:,:,i] = this_trial_sf;

    
```

```python
# create a dict and then a json...
sf_prob_dict = {};

for trial_idx in range(n_trials):
# create a dict for each 
    trial_dict = {}
    for state_idx in range(3):
        trial_dict["s_{}".format(state_idx)] = list(state_feature_probs[state_idx,:,trial_idx])
    sf_prob_dict[trial_idx] = trial_dict

with open('sf_prob.json', 'w') as outfile:
    json.dump(sf_prob_dict, outfile)
```
